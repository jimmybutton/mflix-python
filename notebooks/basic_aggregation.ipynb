```python
unwind_stage = { "$unwind": "$directors" }

group_stage = {
    "$group": {
        "_id": {
            "director": "$directors"
        },
        "average_rating": { "$avg": "$imdb.rating" },
        "num_movies": { "$sum": 1}
    }
}

sort_stage = {
    "$sort": { "average_rating": -1 }
}

# create pipeline from four different stages
pipeline = [
    unwind_stage,
    group_stage,
    sort_stage
]

# aggregate using pipeline
director_ratings = movies.aggregate(pipeline)

# iterate through the resulting cursor
list(director_ratings)
```

```python
skip_stage = {"$skip": 6000}

limit_stage = {"$limit": 100}

facet_stage = {
    "$facet": {
        "runtime": [{
            "$bucket": {
                "groupBy": "$runtime",
                "boundaries": [0, 60, 90, 120, 180],
                "default": "other",
                "output": {
                    "count": {"$sum": 1}
                }
            }
        }],
        "rating": [{
            "$bucket": {
                "groupBy": "$metacritic",
                "boundaries": [0, 50, 70, 90, 100],
                "default": "other",
                "output": {
                    "count": {"$sum": 1}
                }
            }
        }],
        "movies": [{
            "$addFields": {
                "title": "$title"
            }
        }]
    }
}

# create pipeline from four different stages
pipeline = [
    skip_stage,
    limit_stage,
    facet_stage
]

# aggregate using pipeline
faceted = movies.aggregate(pipeline, allowDiskUse=True)

# iterate through the resulting cursor
list(faceted)
```
